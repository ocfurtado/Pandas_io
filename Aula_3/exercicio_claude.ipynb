```python
import pandas as pd
import json
```

```python
with open('cotacoes_fornecedores.json', 'r') as file:
    dados = json.loads(file.read())
```

```python
df_normalizado = pd.json_normalize(dados,
                  record_path='itens',
                  meta=[
                      'data_cotacao', 
                      ['fornecedor', 'codigo'], 
                      ['fornecedor', 'razao_social'], 
                      ['fornecedor', 'cidade'], 
                      ['fornecedor', 'uf']],
                      sep='_')
```

```python
df_normalizado
```

```python
df_normalizado['data_cotacao'] = pd.to_datetime(df_normalizado['data_cotacao'])
```

```python
df_normalizado.dtypes
```

```python
df_normalizado['valor_total'] =( df_normalizado['quantidade'] * df_normalizado['preco_unitario']).round(2)
```

```python
df_normalizado
```

```python
df_normalizado.to_json('cotacoes_consolidado.json', orient='records', force_ascii=False)
```

```python
df_normalizado.to_excel('cotacoes_consolidado.xlsx', sheet_name='cotacoes')
```
